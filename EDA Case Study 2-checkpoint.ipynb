## Data Quality Check and Dealing with the Missing Values:
#### Finding the Percentage of Missing Values in all the Columns:

```python
round(100 * df.isnull().sum()/df.shape[0], 2).sort_values(ascending=False)
```

#### Removing columns with more than 50% missing values:

```python
df = df.loc[:, round(100 * df.isnull().sum()/df.shape[0], 2) <50]
```

#### There are few more columns which are having around 47% missing values. Since these are near to 50%, we are removing these columns as well:

```python
df = df.loc[:, round(100 * df.isnull().sum()/df.shape[0], 2) <47]
```

#### Now checking the shape of the dataset we changed:

```python
df.shape
```

#### Filtering columns with missing percentage values less than 13% and then creating a new DataFrame "less_miss_df":

```python
less_than_13_missing = df.loc[:, (round(100 * df.isnull().sum()/df.shape[0], 2) <14) & round(100 * df.isnull().sum()/df.shape[0], 2)>0]
```

#### Getting data types of colums:

```python
less_than_13_missing.dtypes
```

#### As we can see above that there is only one column "NAME_TYPE_SUITE" with data type as 'object', we can replace missing values with the help of mode of the column:

```python
df['NAME_TYPE_SUITE'] = df['NAME_TYPE_SUITE'].fillna(df['NAME_TYPE_SUITE'].mode()[0])
```

#### Now it will be showing 0 null values in the column "NAME_TYPE_SUITE":

```python
df['NAME_TYPE_SUITE'].isnull().sum()
```

#### Now, treating all the missing values for continuos columns which are filterred with less percentage of missing values:

Selecting all the columns which are with continuous data type:

```python
cont_cols = list(less_than_13_missing.select_dtypes(include = np.number).columns)
```

#### As per the abobe description, few columns are with outliers. Treating the Outliers:
So, we are selecting 3 of them as 'OBS_30_CNT_SOCIAL_CIRCLE', 'OBS_60_CNT_SOCIAL_CIRCLE', 'AMT_REQ_CREDIT_BUREAU_QRT' and replacing all the null values with Median:

```python
df['OBS_30_CNT_SOCIAL_CIRCLE'] = df['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(df['OBS_30_CNT_SOCIAL_CIRCLE'].median())
df['OBS_60_CNT_SOCIAL_CIRCLE'] = df['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(df['OBS_60_CNT_SOCIAL_CIRCLE'].median())
df['AMT_REQ_CREDIT_BUREAU_QRT'] = df['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(df['AMT_REQ_CREDIT_BUREAU_QRT'].median())
```

Now, selecting 2 columns as 'AMT_GOODS_PRICE', 'EXT_SOURCE_2' which are not with outliers and replacing null values with mean of the column:

## Univariate Analysis for Categorical Variables: Target 0 and Target 1:

```python
def uni_categorical_plot(col1,col2):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    sns.countplot(col1,order=col1.value_counts(ascending=False).index)
    plt.xticks(rotation=90)
    plt.title("Target=0")
    plt.subplot(1,2,2)
    sns.countplot(col2,order=col2.value_counts(ascending=False).index)
    plt.xticks(rotation=90)
    plt.title("Target=1")
    plt.show()
```

2. Education Type.
3. Income Type.
4. Age.
5. Family Status.
6. Days employed.
7. Own Realty.
8. Count of children.
